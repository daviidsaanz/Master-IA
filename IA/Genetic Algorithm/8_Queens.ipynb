## **1.Definimos el tamaño del tablero y la pobación**

La **población** equivale al **NUMERO DE TABLEROS** con sus respectivas reinas.
Las **reinas** son representadas con numeros.
La posicion de la reina en la lista representa la columna.
*EX: [1,3,5,2,4]. La reina "1" esta en la columna 1 del tablero y en la posicion 1 (altura). La reina "3" estara en la segunda columna pero en la posicion 3 (altura)*

```python
import random
import math
```

```python
def init_population(mida_tauler, mida_poblacio):
  return [init_individu(mida_tauler) for x in range(mida_poblacio)]
```

```python
def init_individu(mida_tauler):
  individu = [i + 1 for i in range(mida_tauler)]
  random.shuffle(individu)
  return individu
```

```python
def crossover(poblacio):
  individus_nous = []
  parelles_poblacio = seleccio_pares(poblacio)
  for p1, p2 in parelles_poblacio:
    individus_nous.extend(reproduccio(p1, p2))
  return individus_nous

assert len(crossover(init_population(4,10))) == 10
```

```python
def seleccio_pares(poblacio):
  parelles = []
  random.shuffle(poblacio)
  for taulell in range(0, len(poblacio), 2):
    parelles.append([poblacio[taulell], poblacio[taulell+1]])
  return parelles
```

```python
def reproduccio(p1,p2):
  fill1 = p1[:len(p1)//2]
  fill2 = p2[:len(p2)//2]

  for q in p2:
    if q not in fill1:
      fill1.append(q)
  for q in p1:
    if q not in fill2:
      fill2.append(q)
  return mutacio(fill1), mutacio(fill2)
```

```python
def mutacio(fill, mut_probabilitat=0.1):
  if random.random() < mut_probabilitat:
    i1, i2 = random.sample(range(len(fill)),2)
    fill[i1], fill[i2] = fill[i2], fill[i1]
  return fill

assert mutacio([1,2,3,4], 1) != [1,2,3,4]
```

```python
def merge_poblacio(poblacio_inicial, individus_nous):
  return poblacio_inicial + individus_nous
```

```python
def fitness(poblacio_netejar):
    resultats = []
    for taulell in poblacio_netejar:
        conflictes = 0
        n = len(taulell)

        for i in range(n):
            for j in range(i + 1, n):
                if abs(taulell[i] - taulell[j]) == abs(i - j):
                    conflictes += 1
        resultats.append(conflictes)

        if conflictes == 0:
            solucio_trobada(taulell)

    poblacio_amb_fitness = []
    for x in range(len(poblacio_netejar)):
        poblacio_amb_fitness.append((resultats[x], poblacio_netejar[x]))
    poblacio_amb_fitness.sort()
    print(poblacio_amb_fitness)
    poblacio_filtrada = seleccio_escollits(poblacio_amb_fitness)
    return poblacio_amb_fitness
```

```python
def seleccio_escollits(poblacio_amb_fitness, pct_millors=0.3, pct_pitjors=0.1, pct_aleatoris=0.6):
    len_total = len(poblacio_amb_fitness)
    if len_total == 0:
        return []

    n_millors = int(len_total * pct_millors)
    n_pitjors = int(len_total * pct_pitjors)
    n_aleatoris = int(len_total * pct_aleatoris)

    excés = (n_millors + n_pitjors + n_aleatoris) - len_total
    if excés > 0:
        n_aleatoris = max(0, n_aleatoris - excés)

    millors = [ind[1] for ind in poblacio_amb_fitness[:n_millors]]
    pitjors = [ind[1] for ind in poblacio_amb_fitness[-n_pitjors:]] if n_pitjors > 0 else []

    resta = [ind[1] for ind in poblacio_amb_fitness[n_millors:len_total - n_pitjors]]

    seleccionats_aleatoris = random.sample(resta, min(n_aleatoris, len(resta))) if resta else []

    seleccionats = millors + pitjors + seleccionats_aleatoris
    random.shuffle(seleccionats)

    return seleccionats
```

```python
solucio_trobada_flag = False

def solucio_trobada(taulell_seleccionat):
    global solucio_trobada_flag
    print(f"Solució trobada. Taulell: {taulell_seleccionat}")
    solucio_trobada_flag = True
```

```python
def algoritme_genetic(mida_tauler=10, mida_poblacio=20):
    global solucio_trobada_flag
    solucio_trobada_flag = False

    poblacio = init_population(mida_tauler, mida_poblacio)
    generacio = 1

    while not solucio_trobada_flag:
        print(f"Generación {generacio}")

        fills = crossover(poblacio)

        poblacio_completa = merge_poblacio(poblacio, fills)

        poblacio_amb_fitness = fitness(poblacio_completa)

        generacio += 1

        if generacio > 100000:
            print("Has arribat al límit de generacions")
            break

algoritme_genetic()
```
